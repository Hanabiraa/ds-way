## Подготовка

```python
# импорт библиотек
import matplotlib.pyplot as plt
import numpy as np
```

## Данные

```python
np.linspace(0, 9, 10)
```

```python
x = np.linspace(1, 9, 1000)
y1 = np.power(x, 0.5)
y2 = np.sin(x)
y3 = np.cos(x)
```

## Создание графиков при помощи [функционального API](https://matplotlib.org/stable/api/index.html#the-pyplot-api)

> matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

```python
# Простая визуализация
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
```

```python
# визуализация с различным форматированием
## параметр label необходим для задания названия функции 
plt.plot(y1, label='$\sqrt{x}$')
plt.plot(y2, label='$\sin{x}$')
plt.plot(y3, label='$\cos{x}$')

# Отображение легенды для графика (иначе говоря отображение лэйблов функций)
plt.legend()
# задание имен осям графика
plt.xlabel('x')
plt.ylabel('f(x)')
# заголовок графика
plt.title('some different f(x)')
```

## Создание графиков с помощью [объектного API](https://matplotlib.org/stable/api/index.html#the-object-oriented-api)


> matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)

```python
# создание подграфиков одного графика (т.е. график будет разделен на разные части. Каждая часть для своего графика)
## figsize - размер графика
## dpi - его четкость

# fig - управление всеми графиками сразу
# axes - для управления каждым графиком отдельно
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 2), dpi=300)

# в данном случае массив axes - первой размерности
ax1, ax2 = axes
ax1.plot(x, label='x')
ax2.plot(y1, label='$\sqrt{x}$')

# общее название всех графиков
fig.suptitle('2 plots')

# задание каждому из графиков название оси x
ax1.set_xlabel('x')
ax2.set_xlabel('f(x)')

# легенда для всех графиков
fig.legend()

# легенда для конкретного графика
ax2.legend()
```

## Виды графиков в matplotlib

> Axes.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

> Axes.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)

> Axes.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=None, data=None)

```python
names = ['a', 'b', 'c']
values = [100, 150, 180]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10), dpi=300)

# в данном случае массив axes двумерный, ввиду этого необходимо использовать кортежи
(ax1, ax2), (ax3, ax4) = axes

ax1.bar(names, values, color=['r', 'g', 'b'])
ax1.set_title('bar')
ax2.scatter(names, values)
ax2.set_title('scatter')
ax3.plot(names, values)
ax3.set_title('plot')
ax4.pie(values)
ax4.set_title('pie')
```
