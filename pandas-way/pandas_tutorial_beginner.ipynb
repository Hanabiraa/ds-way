## Changing Data

```python
# new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

# new column
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head()
```

```python
df.sort_values('Total', ascending=False)
```

```python
# drop column
df = df.drop(columns='Total')
```

```python
df.head(5)
```

```python
# reordering columns
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]]+cols[4:12]]
```

```python
df.head(5)
```

## Saving Data

```python
# csv
df.to_csv('./data/pokemons/mod_csv_for_pokemons.csv', index=False)
```

```python
# xlsx
df.to_excel('./data/pokemons/mod_csv_for_pokemons.xlsx', index=False)
```

```python
# txt
df.to_csv('./data/pokemons/mod_csv_for_pokemons.txt', index=False, sep='\t')
```

## Filtering Data

```python
# filter by Pandas syntax
df.loc[((df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')) | ((df['Generation'] == 5) * (df['HP'] > 160))]
```

```python
# complex filter
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```
